```python
import numpy as numpy
import pandas as pd
import json
from rank_bm25 import BM25Okapi
import nltk
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv('../scrape/supermarkets.csv')
df
```

```python
items = df['name'].astype(str).tolist()
items
```

```python
nltk.download("punkt")

# Tokenize documents
tokenized_corpus = [word_tokenize(doc.lower()) for doc in items]

# Initialize BM25
bm25 = BM25Okapi(tokenized_corpus)
```

```python
query = "white bread"
tokenized_query = word_tokenize(query.lower())

# Get BM25 scores for all items
scores = bm25.get_scores(tokenized_query)

top_n = 5
top_indices = sorted(range(len(scores)), key=lambda i: scores[i], reverse=True)[:top_n]

# Convert results to JSON array
results = [
    {"rank": rank + 1, "text": items[idx]}
    for rank, idx in enumerate(top_indices)
]

print(results)  # Prints JSON output
```
