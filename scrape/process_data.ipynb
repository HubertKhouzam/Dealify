```python
import pandas as pd
```

```python
maxi = pd.read_csv('maxi.csv')
iga = pd.read_csv('iga.csv')
loblaws = pd.read_csv('loblaws.csv')
provigo = pd.read_csv('provigo.csv')
```

```python
maxi['store'] = 'maxi'
iga['store'] = 'iga'
loblaws['store'] = 'loblaws'
provigo['store'] = 'provigo'
```

```python
maxi.columns
```

```python
maxi['nominal_price'] = maxi['Price'].str.extract(r'(\d+\.\d{2})').astype(float)
iga['nominal_price'] = iga['Price'].str.extract(r'(\d+\.\d{2})').astype(float)
loblaws['nominal_price'] = loblaws['Price'].str.extract(r'(\d+\.\d{2})').astype(float)
provigo['nominal_price'] = provigo['Price'].str.extract(r'(\d+\.\d{2})').astype(float)
```

```python
maxi = maxi.rename(columns={'Price': 'price', 'Name': 'name'})
iga = iga.rename(columns={'Price': 'price', 'Name': 'name'})
loblaws = loblaws.rename(columns={'Price': 'price', 'Name': 'name'})
provigo = provigo.rename(columns={'Price': 'price', 'Name': 'name'})
```
